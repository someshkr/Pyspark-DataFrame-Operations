```python
text =['A big black for jump', 'over the small white dog'] 
rdd = sc.parallelize(text)  
word = rdd.flatMap(lambda line : line.split(" ")).collect()
print(word)
```

```python
def wordFinder(s):
  for l in s:
    if 'for' not in l:
      yield 1
    else:
      yield 0

print(rdd.mapPartitions(wordFinder).collect())
```

```python
data = [1,2,3,4,5]
rdd = sc.parallelize(data,4)

def sumSpecial(splitIndex,iterator):
  if splitIndex%2 ==0:
    yield splitIndex

print(rdd.mapPartitionsWithIndex(sumSpecial).collect())
print(rdd.mapPartitionsWithIndex(sumSpecial).sum())
```

```python
rdd = sc.parallelize(data)
rdd.sample(withReplacement=True, fraction=0.5, seed=45).collect()
```

```python
data1 = [("Java", "20000"), ("Python", "100000"), ("Scala", "3000")]

data2 = [("Java", "200"), ("Python", "100"), ("Scala", "30")]

rdd1  = sc.parallelize(data1)
rdd2  = sc.parallelize(data2)

print(rdd1.union(rdd2).collect())
```

```python
rdd  =sc.parallelize([1,2,3,2,3])
print(rdd.distinct().collect())
```

```python
rdd  =sc.parallelize([(1,3),(3,5),(2,4),(3,2),(3,3)])
print(rdd.groupByKey().mapValues(lambda x: [a+a for a in x]).collect())
```

```python
print(rdd.reduceByKey(lambda x,y: x/y).collect())
```

```python
rdd = sc.parallelize([("a",3),("a",1),("b",7),("a",5)],5)
print(rdd.aggregateByKey(set(), lambda x,y:x|{y}, lambda x, y: x|y).collect())
```

```python
rdd.sortByKey(ascending=True).collect()
```
